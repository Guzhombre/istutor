This exercise is decievingly simple looking, but involves a range of skills such as accurate resampling, understanding of Beer-Lambert transmission, model optimization and fitting.  We will perform a wavelength calibration based on an example transmittance curve from the ATREM radiative transfer model.  I have provided top-of-atmosphere reflectances over Ivanpah playa, a smooth target, to get us started.  We first load the relevant files.  Note that wavelength data is supplied at extremely high spectral resolution.  Plot it!

```python
import numpy as np
import scipy.ndimage.filters as filters
from scipy.linalg import norm, inv
import pylab as plt
import figure_style
from scipy.optimize import minimize
from IPython.core.pylabtools import figsize

# Load the target 
wl = np.loadtxt('data/initial_wavelengths.txt') / 1000.0 # translate to microns
fwhm = np.loadtxt('data/initial_fwhm.txt') / 1000.0 # translate to microns

toa = np.loadtxt('data/20150927_Ivanpah_v1i_toa_1.txt')
transmission = np.loadtxt('data/TRAN_STD_RES_0p1nm_3p0H2O')[:,1]
wl_transmission = np.loadtxt('data/TRAN_STD_RES_0p1nm_3p0H2O')[:,0] / 1000.0 # translate to microns
```

We next define the spectral response function, calculated over wavelengths x, for a multivariate Gaussian with mean mu and standard deviation sigma.

```python
def srf(x, mu, sigma):
     u = (x-mu)/abs(sigma)
     y = (1.0/(np.sqrt(2.0 * np.pi)*abs(sigma)))*np.exp(-u*u/2.0)
     return y/y.sum()
```

Let's try it out by plotting the response of the 85th channel.  Note that the instrument responses are typically given in Full Width at Half Maximum, which is about 2.3548 standard deviations of a Gaussian distribution.

```python
figsize(16,16/8)

wavelength = wl[85]
width = fwhm[85]
response = srf(wl_transmission, wavelength, width/2.3548)

plt.plot(wl_transmission, response,'k')
plt.xlim([0.70,0.850])
plt.xlabel('Wavelength (microns)')
plt.ylabel('Response')
plt.savefig('response.png');
```

Note that wavelength data is supplied at extremely high spectral resolution. Plot it and save to a file!  We boost the default figure size for more resolution.  We can simulate the instrument response of this transmission spectrum by calculating the response function for each channel.  

```python
figsize(16,16/4)
plt.plot(wl_transmission, transmission)
plt.xlim([0.74,0.79])

# Calculate spectral responses for all channels
responses = []
for wavelength, width in zip(wl, fwhm):
    response = np.sum(srf(wl_transmission, wavelength, width/2.3548) * transmission.T)
    responses.append(response)

# Plot the result
plt.plot(wl, responses, 'k')
plt.xlabel('Wavelength (microns)')
plt.xlim([0.70,0.850])
plt.ylim([0,1])
plt.ylabel('Trasmission');
plt.legend(('Transmission','Measurement'))
plt.savefig('transmission.png');
```

We will define two more critical helper functions for this analysis: 

1. A "forward model"that computes a synthetic spectrum with a linear continuum attenuated by the absorption feature, with five free parameters: the continuum offset and slope, the absorption feature depth, the wavelength shift, and a scaling factor on the original spectral full width at half maximum.  
2. An error function to be minimized, which calculates the mismatch between the instrument measurement and the model.

```python
# Fit the TOA spectrum and model 
# uses a state vector x where:
#  x[0] = vertical offset 
#  x[1] = slope 
#  x[2] = depth of absorption feature [0,1] 
#  x[3] = wavelength shift 
#  x[4] = multiplier factor on FWHM
def synth_spectrum(x, toa, toa_wl, model, model_wl, fwhm, b):
    offset, slope, depth, shift, dfwhm = x
    toa_wl_new = (toa_wl+shift)
    use = np.logical_and(toa_wl_new>=b[0], toa_wl_new<=b[1])
    wl = toa_wl_new[use]
    wdelta = model_wl[1]-model_wl[0]
    sigma = fwhm*dfwhm/2.3548/wdelta
    if sigma<=1.0:
        sigma=1.0
    
    # create spectrum, apply instrument response function and resample
    model = np.exp(-depth*np.log(model)) * offset+slope*(model_wl-b[0])
    model_smooth = filters.gaussian_filter1d(model, sigma)
    model_out = np.interp(wl, model_wl, model_smooth)
    return wl, toa[use], model_out

# RMSE error 
def err(x, toa, toa_wl, model, model_wl, fwhm, b):
        wl,obs,mdl = synth_spectrum(x, toa, toa_wl, model, model_wl, fwhm, b)
        er = np.sqrt(pow(mdl-obs,2).mean())
        return er
```

Pick a relevant atmospheric feature window, such as the oxygen A band at 0.75 microns. Find appropriate continuum limits for this feature, and then define an initial state vector solution that puts you in the right ballpark for the free parameters.  Then, find the state vector to minimize error using a standard nonlinear optimization approach such as scipy's optimize.minimize()  

```python
# Do the calibration.  First establish the limits of our fitting window
lims = (0.74,0.78)
args = toa, wl,  transmission, wl_transmission, fwhm[0], lims

# Initial guess of parameters
x_initial = 0.45,0,-0.8,0,0.9

# Optimize to find the best-fitting parameters
res = minimize(err, x_initial, args=args, method='Nelder-Mead')
best = res.x
```

Synthesize the predicted observation using the best-fitting spectral calibration.  Plot the resulting solution against the actual measurement inside the fitting window.

```python
wln, meas, mdl = synth_spectrum(best, toa, wl, transmission, wl_transmission, fwhm[0], lims)

figsize(9,9/1.618)
plt.plot(wln,meas,'ko')
plt.plot(wln,mdl)
plt.legend(('measured','model'))
plt.title('Wavelength shift: %f microns'%res.x[3])
plt.xlabel('Wavelength (microns)')
plt.ylabel('TOA reflectance');
plt.savefig('shift.png')
```
