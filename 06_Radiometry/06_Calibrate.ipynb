This exercise goes back to the basics - raw digital numbers streaming off the focal plane. We apply a calibration solution to get radiance data, identify a location in a scene where in-situ reference data was acquired, and compare the remote measurement to two different predictions based on different Radiative Transfer Models (RTMS).  We load the raw digital numbers and calibration files.  We will use the numpy "memory map" functionality to read the binary file efficiently from disk.

```python
import numpy as np
from scipy.linalg import norm, inv
import pylab as plt
import figure_style
import os.path

# Load the wavelengths
channel, wavelengths, FWHM = np.loadtxt(os.path.join('data','wavelengths.txt')).T

# Convert from microns to nanometers
wavelengths = wavelengths * 1000.0

# Load the data - BIL interleave
rows, cols, bands = 400,598,425
mm = np.memmap('data/dn', dtype=np.uint16, mode='r', shape=(rows,bands,cols))
dn = np.asarray(mm).copy()

# Load the Flat Field (FF)
mm = np.memmap('data/flatfield', dtype=np.float32, mode='r', shape=(bands,cols))
ff = np.asarray(mm).copy()

# Load the dark frame
mm = np.memmap('data/dark', dtype=np.float32, mode='r', shape=(bands,cols))
dark = np.asarray(mm).copy()

# Load the radiometric calibration coefficients
rcc = np.loadtxt('data/rcc.txt')

# Load ATREM radiance prediction
atrem = np.loadtxt(os.path.join('data','rdn_pred_atrem.txt'))
modtran = np.loadtxt(os.path.join('data','rdn_pred_modtran.txt'))
```

Now, apply the calibration solution.  Recall our simple linear radiometric calibration model, where $L$ is radiance, $R_{cc}$ are linear calibration coefficients, $R_{flat}$ are flat field perturbations, and $\alpha_{dark}$ represents the dark offset:

$$ DN = L  R_{cc}^{-1} R_{flat}^{-1} + \alpha_{dark}$$

Calculate the calibrated radiance cube.  Cycle through each row, and apply the dark offset, Radiative Calibration Coefficients, and Flat Field.

```python
rdn = np.zeros((rows,bands,cols), dtype=np.float32)
for i in range(rows):
    rdn[i:i+1] = ((np.squeeze(dn[i:i+1,:,:]) - dark).T * rcc).T * ff
```

Plot the resulting radiance, and the residuals against the ATREM and MODTRAN model solutions.  

<b> Question 1: </b> Which gives you the better match? 
    
<b> Question 2: </b> Is there structure in the residual error, and if so, why?

```python
plt.plot(wavelengths, np.squeeze(rdn[231,:,267]),'k')
plt.plot(wavelengths, np.squeeze(rdn[231,:,267])-atrem.T, 'r')
plt.plot(wavelengths, np.squeeze(rdn[231,:,267])-modtran.T, 'b')
plt.legend(('Measurement','ATREM Residual','MODTRAN Residual'));
plt.xlabel('Wavelength (nm)')
plt.ylabel('Radiance, residual')
```
