This exercise walks you through linear unmixing of remote imaging spectroscopy data using a spectral library.  We import dependencies, and load the wavelengths, reflectances and library of endmembers.  Here the spectra are provided as lists rather than images.

```python
import scipy as s
import pylab as plt
import figure_style
from scipy.optimize import nnls

# Load the data - first column is the wavelength calibration
rfl = s.loadtxt('data/Citrus_Identification_Data.txt', skiprows=1, delimiter=',')[:,3:]
wavelengths = s.loadtxt('data/Citrus_Identification_Wavelengths.txt')
library = s.loadtxt('data/endmembers.txt')
```

Unmix each spectrum in turn using the full library.  Save a representative example of the measured spectrum, the model, and the best-fitting components.

```python
example = 999
for i in range(rfl.shape[0]):
    
    # Unmix! the coefficient vector gives us the areal occurrance fractions for each
    coeff, resid = nnls(library, s.squeeze(rfl[i,:]))
    
    # Plot the example result 
    if i==example:
        
        plt.plot(wavelengths, rfl[i,:])
        plt.plot(wavelengths, library.dot(coeff[:,s.newaxis]))
        for c in range(coeff.size):
            if coeff[c]>1e-6:
                plt.plot(wavelengths,library[:,c]*coeff[c],'k--')
        plt.xlim((400,2500))
        plt.xlabel('Wavelength (nm)')
        plt.ylabel('Reflectance')
        plt.legend(['Measured spectrum','Model','Components'])
        plt.savefig('results/fit.png');
```
